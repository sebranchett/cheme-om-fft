```python
from scipy.fft import fft, ifft, fftfreq, rfft, irfft
import numpy as np
import matplotlib.pyplot as plt
```

```python
def list_print(series):
    for element in series: print(element)
```

```python
x = np.array([1.0, 2.0, 1.0, -1.0, 1.5])
list_print(x)
```

```python
y = fft(x)
list_print(y)
```

```python
yinv = ifft(y)
list_print(yinv)
```

```python
np.sum(x)
```

```python
# Number of sample points
N = 600
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N, endpoint=False)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf = fft(y)
xf = fftfreq(N, T)[:N//2]
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()
```

```python
plt.plot(x, y)
plt.show()
```

```python
x = np.array([1.0, 2.0, 1.0, -1.0, 1.5])
list_print(x)
```

```python
xrfft = rfft(x)
list_print(xrfft)
```

```python
ixrfft = irfft(xrfft, n=len(x))
list_print(ixrfft)
```

## Check what happens if x is complex

```python
x = np.array([1.0 +0.j, 2.0 +0.j, 1.0 +0.j, -1.0 +0.j, 1.5 +0.j])
list_print(x)
```

```python
y = fft(x)
list_print(y)
```

```python
iy = ifft(y)
list_print(iy)
```

```python
yr = rfft(x)
list_print(yr)
```
